# Percentage of accessions viewed in web browsers at uniprot.org in Q1 2021

### Google Analytics Data for uniprot.org

Obtained the following data by running the following query in Google Analytics:

```
Bot filtering profile

AND

Page path level 1 contains:
uniprot

AND

Page path level 2 match:
^(?i)/([OPQ][0-9][A-Z0-9]{3}[0-9]|[A-NR-Z]([0-9][A-Z][A-Z0-9]{2}){1,2}[0-9])(-[0-9]+)?$
```

````python
visited_accessions = set(jan['Page path level 2']) | set(feb['Page path level 2']) | set(mar['Page path level 2'])
````

````python
len(visited_accessions)
````

### Number of all accessions
The following number was obtained by running a `*` query in UniProtKB (on 2021-03-31) and downloading the results as list file and counting the number of lines in this file:

````python
with open('n-uniprot-accessions.txt') as f:
    n_total_accessions = int(f.readline().split()[0])
````

### Percentage of accessions viewed in Q1 2021
The percentage of total UniProtKB accessions that were accessed in the first three months of 2021 on uniprot.org in a web brwoser is:

````python
100 * len(visited_accessions) / n_total_accessions
````
